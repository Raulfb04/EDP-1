```python
!pip install tabulate
import tabulate
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generador (a, c, m, x0):
    x1 = (a * x0 + c) %m
    return x1
```

```python
n = 150
a = 13
c = 45
m = 128
x0 = 1

muestra = []
datos = []

for i in range(n):
    nuevo = generador(a, c, m, x0)
    muestra.append([i, x0, nuevo/128])
    datos.append(x0/128)
    x0 = nuevo
encabezados = ["i", "x_i", "x_{i+1}"]
print(tabulate.tabulate(muestra, headers = encabezados, tablefmt = "plsql", numalign = "center"))
```

```python
media = sum(datos) / len(datos)
print("Media de la muestra:", media)
```

```python
plt.figure(figsize=(8,5))
plt.hist(datos, bins=15, edgecolor='black')
plt.title("Histograma de la muestra")
plt.xlabel("Valores aleatorios")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
