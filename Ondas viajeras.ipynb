```python
# Importamos librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# Parámetros:
# c: velocidad de la onda
# L: longitud del dominio
# x: puntos en el espacio
# t_vals: tiempos en los que evaluaremos la onda

c = 1
L = 1
x = np.linspace(0, L, 500)
t_vals = [0, 0.1, 0.2, 0.3, 0.4]
```

```python
# Definimos la función gaussiana que representa la perturbación
# Si direccion = '+', la onda se mueve a la derecha (x - ct)
# Si direccion = '-', la onda se mueve a la izquierda (x + ct)
def gauss(x, t, direccion='+'):
    if direccion == '+':
        x_prime = x - c * t
    else:
        x_prime = x + c * t
    return np.exp(-((x_prime - 0.5)**2) / 0.01)
```

```python
# Graficamos:
# f = onda que va a la derecha
# g = onda que va a la izquierda
# u = superposición f + g
# Mostramos cómo se propagan con el tiempo
plt.figure(figsize=(12, 6))

for i, t in enumerate(t_vals):
    plt.subplot(2, 3, i+1)
    f = gauss(x, t, '+')
    g = gauss(x, t, '-')
    u = f + g

    plt.plot(x, f, 'b--', label='Derecha')
    plt.plot(x, g, 'r--', label='Izquierda')
    plt.plot(x, u, 'k', label='Superposición')
    plt.title(f"t = {t:.2f}")
    plt.ylim(0, 2.1)
    plt.grid(True)
    if i == 0:
        plt.legend()

plt.tight_layout()
plt.show()
```
