En esta parte calculamos la media muestral

```python
x=sum(data)/len(data)
print("media muestral=", x)
```

Obtenemos z0

```python
z0= ((x-0.5) * sqrt(len(data)))/sqrt(1/12)
print("z0=", z0)
```

Obtenemos el valor crtitico

```python
alpha = 0.05
z_critico = norm.ppf(1- alpha/2)
print("z_critico=", z_critico)
```

Hacemos la prueba de hipotesis

```python
if abs(z0) > z_critico:
    print("Rechazar H0")
else:
    print("No rechazar H0")
```
